```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Insertion des donn√©es dans la variable voitures

```python
voitures = pd.read_csv("datasets_sell_car.csv",sep=",")
```

### Affichage de la variable voitures

```python
voitures
```

### Affichage du nombre total de model Audi vendu en millier

```python
voitures[voitures.Manufacturer=="Audi"].Sales_in_thousands.sum()
```

### Classement des voitures vendu des plus on√©reuse au moins ch√®re

```python
voitures.sort_values(by="Price_in_thousands", ascending = False)
```

### Les mod√®les de voitures qui consomment le moins

```python
voitures.sort_values(by="Fuel_efficiency")
```

### Pour chaque constructeur l'affichage du total des ventes

```python
voitures.groupby("Manufacturer").Sales_in_thousands.sum()
```

### Nouveau tableau permettannt de voir pour chaque marque la moyenne du facteur de performance pour les types car et passenger

```python
average_performance=voitures.pivot_table('Power_perf_factor',index='Manufacturer',columns='Vehicle_type',aggfunc=np.mean)
average_performance
```

```python
performance=voitures.pivot_table('Power_perf_factor',index ='Horsepower',columns='Manufacturer',aggfunc=np.mean)
performance.plot()
```
